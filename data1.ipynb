```python
new_df = df[['Base_Type', 'Product_Type','Legislation']]
```

```python
new_df
```

```python

# Prepare new_df for Gower: keep categorical cols as object, not 'category'
new_df2 = new_df.copy()
cat_cols = new_df2.select_dtypes(include=['object','category']).columns.tolist()
for c in cat_cols:
    new_df2[c] = new_df2[c].fillna('__MISSING__').astype(object)

# Tell gower which columns are categorical
cat_features = np.array([col in cat_cols for col in new_df2.columns])

# Full pairwise Gower (n x n)
D = gower.gower_matrix(new_df2, cat_features=cat_features)
D_df = pd.DataFrame(D, index=new_df.index, columns=new_df.index)

# Distances from the first row by position (row 0)
row_pos = 0
row_idx = new_df.index[row_pos]
dists_row0 = (
    D_df.loc[row_idx]
    .drop(row_idx)
    .sort_values()
)

# Top-k nearest to row 0
k = 50
topk_idx = dists_row0.index[:k]
topk_rows = df.loc[topk_idx].assign(gower_distance=dists_row0.values[:k])
topk_rows
```
